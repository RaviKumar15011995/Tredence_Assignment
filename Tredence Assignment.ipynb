```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from fuzzywuzzy import fuzz 
from fuzzywuzzy import process 
import warnings
warnings.filterwarnings('ignore')


manuf_data=pd.read_excel(r'D:\Tredence\manuf_data.xlsx')
retailer_data=pd.read_excel(r'D:\Tredence\retailer_data.xlsx')
manuf_data_dic={}
retailer_data_dic={}
```

```python
#Creating the dictionary where key=firstcolumn i.e ID , value=secondcolumn i.e description in both datasets
manuf_data_dic={ row[0]:row[1]  for index,row in manuf_data.iterrows()}
retailer_data_dic={ row[0]:row[1]  for index,row in retailer_data.iterrows()}
```

```python
#Removing Stop words and Finding similarity match between each retailer data with manufacture data

for keyR in retailer_data_dic:
    ret_text=[i for i in str(retailer_data_dic[keyR]).split() if not i in  stopwords.words('english')]
    ret_text_join=''.join(ret_text)
    similarity_retail_match=[]
    
    for keyM in manuf_data_dic:
        manuf_text=[i for i in str(manuf_data_dic[keyM]).split() if not i in  stopwords.words('english')]
        manuf_text_join=''.join(manuf_text)
        percent=fuzz.ratio(manuf_text_join,ret_text_join)
        #print(keyR,keyM)
        cord=[keyM,percent]
        similarity_retail_match.append(cord)
    similarity_retail_match=sorted(similarity_retail_match,key=lambda x:-x[1])
    
    similarity_retail_match=similarity_retail_match[0:4]
    
    
    # Finding each retailer id four most similar manufacture id by looking into description
    print('Top 4 match of each retailer id with manufacture id by looking into description\n')
    for key,value in similarity_retail_match:
        
        print('Retailer ID','  ','Manufacture ID')
        print(keyR,'             ',key)
    print('\n\n')
            
```
